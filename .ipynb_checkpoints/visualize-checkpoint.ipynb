```python
import numpy as np
import pickle
import matplotlib.pyplot as plt

def load(fp):
    with open(fp, "rb") as p:
        return pickle.load(p)

def plot_loss_history(title, ax, loss_history, fmt=".", limit=None, label=None):
    ax.set_ylabel('J(Theta)')
    ax.set_xlabel('Iterations')
    ax.set_title(title)
    
    if limit is not None:
        loss_history = loss_history[: limit]
        
    _=ax.plot(range(len(loss_history)),loss_history, fmt, label=label)
    
    ax.legend()
    
# def plot_multiple_loss_history(title, result_set, limit=None):
#     result_keys = result_set.keys()
#     result_values = result_set.values()
#     len_result = len(result_keys)
    
#     fig,ax = plt.subplots(figsize=(12,8))
    
#     for (label, result) in zip(result_keys, result_values):
#         plot_loss_history(title, ax, result, limit=limit, label=label)
```
