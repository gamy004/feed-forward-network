```python
import numpy as np
import pickle
import matplotlib.pyplot as plt

def load(fp):
    with open(fp, "rb") as p:
        return pickle.load(p)

def plot_loss_history(title, ax, loss_history, fmt=".", limit=None, label=None):
    ax.set_ylabel('J(Theta)')
    ax.set_xlabel('Iterations')
    ax.set_title(title)
    
    if limit is not None:
        loss_history = loss_history[: limit]
        
    _=ax.plot(range(len(loss_history)),loss_history, fmt, label=label)
    
    ax.legend()
```

```python
model = load('./model.sav')

training_loss_count = len(model.L)
val_loss_count = len(model.V_L)
val_acc_count = len(model.V_ACC)

print("Total loss history: {}".format(training_loss_count))
```

```python
fig,ax = plt.subplots(figsize=(12,8))

plot_loss_history('Training Loss VS Validate Loss',
                  ax,
                  model.L,
                  label='Training Loss'
                 )

plot_loss_history('Training Loss VS Validate Loss',
                  ax,
                  model.V_L,
                  label='Validate Loss'
                 )
```

```python
fig,ax = plt.subplots(figsize=(12,8))

plot_loss_history('Training Loss VS Validate Loss between round 1-1000',
                  ax,
                  model.L,
                  limit=1000,
                  label='Training Loss'
                 )

plot_loss_history('Training Loss VS Validate Loss between round 1-1000',
                  ax,
                  model.V_L,
                  limit=1000,
                  label='Validate Loss'
                 )
```
